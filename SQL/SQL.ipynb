## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

- Câu 12:
    Mysql:
        select T.team_long_name, sum(goal) as total_goal
        from (
            select home_team_api_id as t_api_id, sum(home_team_goal) as goal from Match group by home_team_api_id 
            UNION ALL 
            select away_team_api_id, sum(away_team_goal) from Match group by away_team_api_id
            ) as N 
        left outer join Team as T on N.t_api_id = T.team_api_id 
        group by T.team_long_name 
        order by sum(goal) DESC

```python
# ĐIỀN CODE VÀO ĐÂY
result12 = pd.read_sql('select T.team_long_name, sum(goal) as total_goal from (select home_team_api_id as t_api_id, sum(home_team_goal) as goal from Match group by home_team_api_id UNION ALL select away_team_api_id, sum(away_team_goal) from Match group by away_team_api_id) as N left outer join Team as T on N.t_api_id = T.team_api_id group by T.team_long_name order by sum(goal) DESC',conn)
result12
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


- Câu 14:
    Mysql:
        select team_long_name, result, count(result) as num_result 
        from (
            select home_team_api_id, CASE 
                when (home_team_goal - away_team_goal) > 0 then "win" 
                when (home_team_goal - away_team_goal) < 0 then "lose" 
                else "draw" 
                end as result 
                from Match
            ) as R left outer join Team as T on R.home_team_api_id = T.team_api_id 
        group by team_long_name, result

```python
# ĐIỀN CODE VÀO ĐÂY
result14 = pd.read_sql('select team_long_name, result, count(result) as num_result from (select home_team_api_id, CASE when (home_team_goal - away_team_goal) > 0 then "win" when (home_team_goal - away_team_goal) < 0 then "lose" else "draw" end as result from Match) as R Left outer join Team as T on R.home_team_api_id = T.team_api_id group by team_long_name, result',conn)
result14
```

#### Câu 14: Em thử tổng hợp lại bảng số bàn thắng, thua, hòa trên 1 dòng

- Câu 14:
    Mysql:
            select team_long_name, sum(win_result) as win_result, sum(lose_result) as lose_result, sum(draw_result) as draw_result 
            from Team as T 
            left outer join (select home_team_api_id, count(result) as win_result from (select home_team_api_id, CASE when (home_team_goal - away_team_goal) > 0 then "win" end as result from Match where result = "win") group by home_team_api_id) as W
            on T.team_api_id = W.home_team_api_id 
            left outer join (select home_team_api_id, count(result) as lose_result from (select home_team_api_id, CASE when (home_team_goal - away_team_goal) < 0 then "lose" end as result from Match where result = "lose") group by home_team_api_id) as L 
            on T.team_api_id = L.home_team_api_id 
            left outer join (select home_team_api_id, count(result) as draw_result from (select home_team_api_id, CASE when (home_team_goal - away_team_goal) == 0 then "draw" end as result from Match where result = "draw") group by home_team_api_id) as D
            on T.team_api_id = D.home_team_api_id 
            group by team_long_name 
            order by team_long_name

##### *Vì có đội trùng tên nhưng khác mã team_api_id nên em dùng hàm sum(win_result) và group by theo team_long_name để gộp kết quả của các đội có tên giống nhau lại*

```python
result14_2 = pd.read_sql('select team_long_name, sum(win_result) as win_result, sum(lose_result) as lose_result, sum(draw_result) as draw_result from Team as T left outer join (select home_team_api_id, count(result) as win_result from (select home_team_api_id, CASE when (home_team_goal - away_team_goal) > 0 then "win" end as result from Match where result = "win") group by home_team_api_id) as W on T.team_api_id = W.home_team_api_id left outer join (select home_team_api_id, count(result) as lose_result from (select home_team_api_id, CASE when (home_team_goal - away_team_goal) < 0 then "lose" end as result from Match where result = "lose") group by home_team_api_id) as L on T.team_api_id = L.home_team_api_id left outer join (select home_team_api_id, count(result) as draw_result from (select home_team_api_id, CASE when (home_team_goal - away_team_goal) == 0 then "draw" end as result from Match where result = "draw") group by home_team_api_id) as D on T.team_api_id = D.home_team_api_id group by team_long_name order by team_long_name',conn)
result14_2
```
